# Import Libraries

```python
import importer
import utility
import os
from datetime import datetime, timezone
import numpy as np
```

# Load X-Band Dual Polarization Radar (PERSER) installed at Monserrato, Sardinia, Italy

```python
# get the root_path accordingly
root_path = os.getcwd()
# Define the data directory for the repository data
data_path = os.path.join(root_path, "Data")
# Declare input parameters
radar_type = "X-Band-dualpol"
data_type = "CZW"
ele_info = "805"
num_prev_files =None
num_next_files=None
scan_interval=5
# datetime information for current timestamp
dt = datetime(2025, 4, 16, 22, 0, 0)  # April 16, 2025 at 22:00:00
# Get the data and metadata
dbz_perser, metadata_perser_dbz = importer.import_files_by_date(data_path,radar_type, data_type, ele_info, dt,scan_interval,num_prev_files,num_next_files)
```

```python
dn_perser,metadata_perser_dn = utility.dbz_to_dn(dbz_perser, metadata_perser_dbz)
# Noise removal using Watershed technique
dn_perser_clean = np.empty_like(dn_perser)
for t in range(dn_perser.shape[0]):
    dn_perser_clean[t, :, :] = utility.noise_remove(dn_perser[t, :, :], type="Watershed")

# Convert Digital Number to Reflectivity (dBZ)
dbz_perser_clean,metadata_perser_dbz = utility.dn_to_dbz(dn_perser_clean,metadata_perser_dn)
```

```python
# Convert to rain rate (mm/h)
from pysteps.utils import conversion
R_perser, metadata_perser_R = conversion.to_rainrate(dbz_perser_clean, metadata_perser_dbz)
```

# Load C-Band Italian Civil Protection Radar data

```python
# Declare input parameter
radar_type = "C-Band"
data_type = "VMI"
ele_info = None
num_prev_files =None
num_next_files= None
scan_interval=5
# datetime information
dt = datetime(2025, 4, 16, 23, 0, 0)  # April 16, 2025 at 23:00:00
# Get the data and metadata
dbz_cp, metadata_cp_dbz = importer.import_files_by_date(data_path,radar_type, data_type, ele_info, dt,scan_interval,num_prev_files,num_next_files)
```

```python
# Convert to rain rate (mm/h)
R_cp, metadata_cp_R = conversion.to_rainrate(dbz_cp, metadata_cp_dbz)
```

# Load X-Band Single polar radar installed at UNICA, Italy

```python
# Declare input parameter
radar_type = "X-Band"
data_type = "data"
ele_info = None
num_prev_files =None
num_next_files=None
scan_interval=5
# datetime information
dt = datetime(2023, 5, 20, 18, 50, 0)  # May 20, 2023 at 18:50:00
# Get the data and metadata
dn_gavino, metadata_gavino_dn = importer.import_files_by_date(data_path,radar_type, data_type, ele_info, dt,scan_interval,num_prev_files,num_next_files)
```

```python
# Noise removal using Watershed technique
dn_gavino_clean = np.empty_like(dn_gavino)
for t in range(dn_gavino.shape[0]):
    dn_gavino_clean[t, :, :] = utility.noise_remove(dn_gavino[t, :, :], type="Watershed")

# Convert Digital Number to Reflectivity (dBZ)
dbz_gavino,metadata_gavino_dbz = utility.dn_to_dbz(dn_gavino_clean,metadata_gavino_dn)
```

```python
# Convert to rain rate (mm/h)
R_gavino, metadata_gavino_R = conversion.to_rainrate(dbz_gavino*1.4, metadata_gavino_dbz)
```

# Find geographical information for Civil protection data needs to be ingested inside pysteps module

```python
import rasterio
# Load the GeoTIFF file
with rasterio.open(r"D:\Geosciences_Project\Nowcasting_OF\Data\Data_Civil_Protection\geodata\VMI_geodata.tif") as src:
    bounds = src.bounds
    x1, y1, x2, y2 = bounds.left, bounds.bottom, bounds.right, bounds.top

    ## Method 1: Using src.transform 
    transform = src.transform
    x_pixel_size = transform.a       # width of a pixel in x direction
    y_pixel_size = -transform.e      # height of a pixel in y direction (usually negative, so take negative)

    ## Method 2: Using bounds and array dimensions
    width = src.width       # number of columns
    height = src.height     # number of rows
    bounds = src.bounds     # spatial extent
    x_pixel_size_1 = (bounds.right - bounds.left) / width
    y_pixel_size_1 = (bounds.top - bounds.bottom) / height

print(f"x1: {x1}, x2: {x2}, y1: {y1}, y2: {y2}")
print(f"x pixel size: {x_pixel_size}")
print(f"y pixel size: {y_pixel_size}")
print(f"x pixel size_1: {x_pixel_size_1}")
print(f"y pixel size_1: {y_pixel_size_1}")
```

# Plotting three types of data

```python
from pysteps.visualization import plot_precip_field
from matplotlib import pyplot as plt
# Disable warnings
import warnings
warnings.filterwarnings("ignore")
fig = plt.figure(figsize=(6, 10), layout="constrained")
# set the title to the timestamp of the last precipitation field
plt.suptitle("Plot Three different types of Radar data available at UNICA")
map_kwargs = {"drawlonlatlines": True}

# plot the last precipitation field
# turn off axis, because it will be overwritten in plot_precip_field
ax = fig.add_subplot(311); ax.axis("off")
plot_precip_field(
    R_perser[0],
    geodata=metadata_perser_R,
    title=f"Perser Radar of {metadata_perser_R['timestamps'][0].strftime('%Y-%m-%d %H:%M:')}",
    colorscale="pysteps",
    map_kwargs=map_kwargs,
    units="mm/h",
    ax=ax,
); # the default colormap
ax = fig.add_subplot(312); ax.axis("off")
plot_precip_field(
    R_cp[0],
    geodata=metadata_cp_R,
    title=f"Civil Protection Radar of {metadata_cp_R['timestamps'][0].strftime('%Y-%m-%d %H:%M:')}",
    colorscale="pysteps",
    map_kwargs=map_kwargs,
    units="mm/h",
    ax=ax,
);
ax = fig.add_subplot(313); ax.axis("off")
plot_precip_field(
    R_gavino[0],
    geodata=metadata_gavino_R,
    title=f"Gavino Radar of {metadata_gavino_R['timestamps'][0].strftime('%Y-%m-%d %H:%M:')}",
    colorscale="pysteps",
    map_kwargs=map_kwargs,
    units="mm/h",
    ax=ax,
);
```
